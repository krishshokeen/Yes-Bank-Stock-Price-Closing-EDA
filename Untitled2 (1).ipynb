# **3. Data Wrangling**

---



### **Data Wrangling Code**

```python

#Convert Date column to Datetime format

df['Date'] = pd.to_datetime(df['Date'], format='%b-%y')
df.sort_values('Date', inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python

# Extract Time-Based Features
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Quarter'] = df['Date'].dt.quarter
```

```python
# Creating Lag Features (Very Important)
df['Close_lag_1'] = df['Close'].shift(1)
df['Close_lag_2'] = df['Close'].shift(2)
df['Close_lag_3'] = df['Close'].shift(3)
```

```python
df[['Close','Close_lag_1','Close_lag_2','Close_lag_3']].head()
```

```python

# Rolling Statistics (Trend & Volatility Capture)
df['Rolling_Mean_3'] = df['Close'].rolling(window=3).mean()
df['Rolling_Mean_6'] = df['Close'].rolling(window=6).mean()
df['Rolling_STD_6'] = df['Close'].rolling(window=6).std()
```

```python
# Price Range & Volatility Features
df['Price_Range'] = df['High'] - df['Low']
df['Range_Percentage'] = (df['High'] - df['Low']) / df['Close'] * 100
```

```python

# Return-Based Features
df['Monthly_Return'] = df['Close'].pct_change()
df['Log_Return'] = np.log(df['Close'] / df['Close'].shift(1))
```

```python
# Handling NA Values Created by Feature Engineering
df.isnull().sum()
```

```python

df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

### **What all manipulations have you done and insights you found**
1. Converted the Date column into datetime format.
2. Checked and confirmed there were no missing values or duplicate records in the original dataset.
3. Created new features such as lag variables, rolling mean, rolling standard deviation, price range, monthly return, and log return.
4. Removed the missing values generated due to feature engineering using dropna() and reset the index.

Insights:

1. The stock price variables (Open, High, Low, and Close) are highly correlated.
2. Historical closing prices and rolling statistics help in predicting future closing prices.
3. The engineered features provide additional information that improves the predictive capability of the model.
